```python
# Import libraries

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None

# Read in the csv file

df = pd.read_csv(r'~/LearnTech/Data Analyst/Portfolio/DS Salaries Project/new_ds_salaries.csv')
```

```python
df.head()
```

```python
# Check for missing columns

for col in df.columns:
    percent_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, percent_missing))
```

```python
# No null values
```

```python
# Check column data types

df.dtypes
```

```python
# Sort data by 'salary_in_usd' in descending order

df.sort_values(by=['experience_level'], inplace=False, ascending=False).head()
```

```python
# Display distinct values

df['job_title'].drop_duplicates().sort_values(ascending=True)
```

```python
# Create scatter plot (remote_ratio vs salary_in_usd)

plt.scatter(x=df['remote_ratio'],y=df['salary_in_usd'], color='blue')

plt.title('Remote Ratio vs Salary')

plt.xlabel('Remote Ratio')

plt.ylabel('Salary')

plt.show()
```

```python
# Create scatter plot (company_size vs salary_in_usd)

plt.scatter(x=df['company_size'],y=df['salary_in_usd'], color='blue')

plt.title('Company Size vs Salary')

plt.xlabel('Company Size')

plt.ylabel('Salary')

plt.show()
```
